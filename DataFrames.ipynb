8. Conditional Selection

Q.1 Find peoples those having salary above 70000

```python
df_arr
```

```python
ans_q1 = df_arr[ df_arr["Salary"] > 70000 ]
```

```python
ans_q1
```

Q.2 Find peoples those having EmployeeID 102

```python
ans_q2 = df_arr[ df_arr["EmployeeID"] == 102 ]
ans_q2
```

Q.3 Find peoples those from Department of Engineering

```python
ans_q3 = df_arr[ df_arr["Department"] == "Engineering" ]
ans_q3
```

Q.4 Find peoples those having salary above 90000 and they are from Department of Engineering

```python
ans_q4 = df_arr[ (df_arr["Salary"] > 90000) & (df_arr["Department"] == "Engineering")]
ans_q4
```
