```python
def extract_eps(text):
    check = False
    data = ""
    for i in text:
        if i == "(":
            check = True
            continue
        if i == ")":
            if check:
                return data
        if check:
            data += i
```

```python
df['Episodes'] = df['Title'].apply(extract_eps)
```

```python
import re

def extract_date_range(text):
    """
    Extracts the date range in the format 'Mon YYYY - Mon YYYY' 
    from a string like: '[Title][Type] ([ep count])Apr 2009 - Jul 2010[members]...'
    """
    try:
        # Step 1: Find the closing parenthesis ')'
        closing_paren_index = text.index(')')

        # Step 2: Take the substring after the closing parenthesis
        after_paren_text = text[closing_paren_index + 1 : closing_paren_index + 50]  # limit range

        # Step 3: Use regex to find the date range
        match = re.search(r'[A-Za-z]{3} \d{4} - [A-Za-z]{3} \d{4}', after_paren_text)

        # Step 4: Return match or handle no match
        if match:
            return match.group()
        else:
            return None  # or "Date range not found"

    except ValueError:
        return None  # or "Closing parenthesis not found"
```

```python
df['Time'] = df['Title'].apply(extract_date_range)
```

```python
import pandas as pd
import re

# Function to extract start month number
def extract_start_month_number(text):
    month_map = {
        'Jan': 1, 'Feb': 2, 'Mar': 3, 'Apr': 4,
        'May': 5, 'Jun': 6, 'Jul': 7, 'Aug': 8,
        'Sep': 9, 'Oct': 10, 'Nov': 11, 'Dec': 12
    }
    match = re.search(r'([A-Za-z]{3}) \d{4} -', text)
    if match:
        return month_map.get(match.group(1))
    return None

# Apply function to create new column
df['Month'] = df['Title'].apply(extract_start_month_number)
```
